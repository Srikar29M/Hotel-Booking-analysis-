```python
def rrt_art_same(df):
  '''
  This function returns whether the customer get their preferred reserved room type or not.
  '''
  if df['reserved_room_type'] == df['assigned_room_type']:
      return 'True'
  else:
      return 'False'
```

```python
df['rrt_art_same']  = df.apply(rrt_art_same ,axis = 1)

df[['is_canceled' , 'reserved_room_type' , 'assigned_room_type' , 'rrt_art_same']]
```

```python
p1 = df['reserved_room_type'].value_counts().reset_index()
p2 = df['assigned_room_type'].value_counts().reset_index()
```

```python

merged_RJ_df = pd.merge(p1, p2, how='right', left_on='reserved_room_type', right_on='assigned_room_type')


merged_RJ_df.fillna(0, inplace=True)


merged_RJ_df = merged_RJ_df.set_index('assigned_room_type')  


print("Merged DataFrame:\n", merged_RJ_df.head())

```

```python
import matplotlib.pyplot as plt


reserved_counts = merged_RJ_df['count_x'] 
assigned_counts = merged_RJ_df['count_y']  

plt.figure(figsize=(20, 10), dpi=110)


plt.plot(reserved_counts.index, reserved_counts, 'go--', linewidth=2, markersize=4, label='Reserved Room')
plt.plot(assigned_counts.index, assigned_counts, 'ro-', linewidth=2, markersize=4, label='Assigned Room')


plt.title('Comparison between Reserved Room and Assigned Room', weight='bold')
plt.xlabel('Room Type')
plt.ylabel('Count')


plt.legend(title='Room Type')


plt.xticks(rotation=45)  
plt.show()
```
